```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
data=pd.read_csv("SPECTRA.txt",sep='\s+',skiprows=0)
data.columns=['y','1.4','10.2','18.3','24.1','27.7','29.1','29.4','29.8','30.1','30.5','r','pos','rbe']
positions=['1.4',"10.2","18.3","24.1","27.7","29.1","29.4","29.8","30.1","30.5"]
data
```

```python
fig1=plt.figure(figsize=(12,4))
a_ax1=plt.subplot2grid((1,2),(0,0),colspan=1,rowspan=1) 
a_ax2=plt.subplot2grid((1,2),(0,1),colspan=1,rowspan=1) 
a_ax1.plot(data[:]['y'],data[:]['30.5'])
a_ax1.set_xscale('log')
for j in positions[5:]:
    m,a=np.polyfit(data[97:103]['y'],data[97:103][j],1)
    data.loc[:97,j]=m*data.loc[:97]['y']+a
a_ax2.plot(data[:]['y'],data[:]['30.5'])
a_ax2.set_xscale('log')
a_ax1.set_title('before extrapolation')
a_ax2.set_title('after extrapolation')
a_ax1.set_xlabel('y[kev/micro m]')
a_ax2.set_xlabel('y[kev/micro m]')
a_ax1.set_ylabel('frequency distribution')
a_ax2.set_ylabel('frequency distribution')
fig1.savefig('extrapolation.png')
```

```python
#for i in range(2):
 #   ax[i]=plt.subplot2grid((2,2),(0,i),colspan=1,rowspan=1)
#ax1=plt.subplot2grid((2,2),(0,0),colspan=1,rowspan=1)
#ax2=plt.subplot2grid((2,2),(0,1),colspan=1,rowspan=1)
integ=np.zeros(10)
for i in range(300):
    for j in range(10):
        integ[j]+=(data.loc[i+1,'y']-data.loc[i,'y'])*(data.loc[i+1,positions[j]]+data.loc[i,positions][j])/2
for j in range(10):
    #print(integ[j])
    data[:][positions[j]]/=integ[j]
```

```python
fig2=plt.figure(figsize=(20,5))
ax=[plt.subplot2grid((2,5),(i//5,i%5),colspan=1,rowspan=1) for i in range(10)]
avg=np.zeros((10,3))
for i in range(300):
    for j in range(10):
        avg[j][0]+=(data.loc[i+1,'y']-data.loc[i,'y'])*(data.loc[i+1,positions[j]]+data.loc[i,positions[j]])*(data.loc[i+1,'y']+data.loc[i,'y'])/4
for j in range(10):
    data[('%sdos'%positions[j])]=data[positions[j]]*data['y']/avg[j][0]
#print(data)
for i in range(300):
    for j in range(10):
        avg[j][1]+=(data.loc[i+1,'y']-data.loc[i,'y'])*(data.loc[i+1,"%sdos"%positions[j]]+data.loc[i,"%sdos"%positions[j]])/2
for j in range(10):
    data[:]["%sdos"%positions[j]]/=avg[j][1]
    avg[j][1]=0
for j in range(10):
    data[('%sydy'%positions[j])]=data["%sdos"%positions[j]]*data['y']
for i in range(300):
    for j in range(10):
        avg[j][1]+=(data.loc[i+1,'y']-data.loc[i,'y'])*(data.loc[i+1,"%sdos"%positions[j]]+data.loc[i,"%sdos"%positions[j]])*(data.loc[i+1,'y']+data.loc[i,'y'])/4 
print(avg)
for j in range(10):
    ax[j].plot(data[:]['y'],data[:][positions[j]])
    ax[j].plot(data[:]['y'],data[:]["%sydy"%positions[j]],color='r')
    ax[j].axvline(x=avg[j][0],ymin=0,ymax=0.25)
    ax[j].axvline(x=avg[j][1],ymin=0,ymax=0.25,color='r')
    ax[j].set_xscale('log')
fig2.savefig('f_and_yd distributions.png')
```

```python
fig3=plt.figure(figsize=(10,5))
b_ax=plt.subplot2grid((1,1),(0,0),colspan=1,rowspan=1)
for j in range(10):
    b_ax.plot(data[:]['y'],data[:]["%sydy"%positions[j]],label=positions[j])
b_ax.set_xscale('log')
b_ax.set_title('ydy')
b_ax.set_xlabel('y [kev/micro m]')
b_ax.set_ylabel('ydy distribution')
b_ax.legend()
fig3.savefig('ydy_distribution.png')
```
